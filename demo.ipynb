This demo shows how to use the custom conditional kernel density estimation sampler. To recreate simply download the nonparametric_kde sampler folder and apply the below to your dataset. 

Note that the sampler can simulate up to 10 variables, and works best with relatively short training datasets. 

```python
import sklearn
import statsmodels
from nonparametric_kde.ckde import ckde
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

1. Read in non-parametric data

```python
iris_data = pd.read_csv('../data/iris.csv')
```

```python
iris_data.iloc[1:600, :3].head()
```

2. Initialize ckde Class Object  

```python
# pass in dataframe of up to 10 variables
# n describes the number of variables you wish to simulate from the dataframe you pulled in. 
iris_ckde = ckde(iris_data.iloc[1:600, :3], n=3)
```

3. Simulate Points 

```python
# takes one argument - number of points to simulate
sim_points = iris_ckde.generate_sim_points(1000)
```

4. Display CKDE Simulation Results and the Capture of Variable to Variable Relationships

```python
fig, (ax2, ax1) = plt.subplots(1, 2, figsize = (10, 4))
ax1.scatter(sim_points['sepal_length'], sim_points['sepal_width'])
ax1.set_title('Simulated Points')
ax1.set_xlabel('Sepal Length')
ax1.set_ylabel('Sepal Width')
ax2.scatter(iris_data['sepal_length'], iris_data['sepal_width'])
ax2.set_title('Training Points')
ax2.set_xlabel('Sepal Length')
ax2.set_ylabel('Sepal Width')
plt.tight_layout()
plt.show()
```

```python
fig, (ax2, ax1) = plt.subplots(1, 2, figsize = (10, 4))
ax1.scatter(sim_points['sepal_length'], sim_points['petal_length'])
ax1.set_title('Simulated Points')
ax1.set_xlabel('Sepal Length')
ax1.set_ylabel('Petal Length')
ax2.scatter(iris_data['sepal_length'], iris_data['petal_length'])
ax2.set_title('Training Points')
ax2.set_xlabel('Sepal Length')
ax2.set_ylabel('Petal Length')
plt.tight_layout()
plt.show()
```

```python
fig, (ax2, ax1) = plt.subplots(1, 2, figsize = (10, 4))
ax1.scatter(sim_points['sepal_width'], sim_points['petal_length'])
ax1.set_title('Simulated Points')
ax1.set_xlabel('Sepal Width')
ax1.set_ylabel('Petal Length')
ax2.scatter(iris_data['sepal_width'], iris_data['petal_length'])
ax2.set_title('Training Points')
ax2.set_xlabel('Sepal Width')
ax2.set_ylabel('Petal Length')
plt.tight_layout()
plt.show()
```
